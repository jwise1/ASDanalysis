```python
data = pd.read_csv("./data/imageDataAnalysismtcnn.csv")
```

```python
for i in data.values:
    fig, axs=plt.subplots(1,2,figsize=(12,6))
    x = data.columns[2:9]
    
    axs[0].bar(x, i[2:9], label='Emotion Probability')
    axs[0].set_title('Confidence Scores of '+i[1])
    img = mpimg.imread(i[1])
    axs[1].imshow(img)
    axs[1].axis('off')
    plt.show()
    #fig.savefig("./data/Graphs/"+str(i[1][11:]).replace('.jpg','-').replace('.JPG','-').replace('Autistic.','')+i[9])
```
